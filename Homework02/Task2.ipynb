```python
import pandas as pd
import numpy as np
import nltk
import warnings 
warnings.filterwarnings("ignore", category=DeprecationWarning)
import os
```

```python
test_df_prepare = pd.read_pickle("output/test_df_processed.pkl")
train_df_prepare = pd.read_pickle("output/train_df_processed.pkl")
test_df_prepare
```

```python
from gensim.models import Word2Vec
```

```python
modelW2V = Word2Vec(sentences=test_df_prepare['tweet_token'],
                    size=200,
                    window=5,
                    min_count=2, 
                    sg=1,
                    hs=0,
                    negative=10,
                    workers=32,
                    seed=34)
```

```python
modelW2V.train(test_df_prepare['tweet_token'],
               total_examples=len(test_df_prepare['tweet_token']),
               epochs=20)
```

```python
modelW2V.save('modelW2V_tweet_token.model')
```

__6. Поскольку наши данные содержат твиты, а не только слова, нам придется придумать способ использовать векторы слов из модели word2vec для создания векторного представления всего твита. Существует простое решение этой проблемы, мы можем просто взять среднее значение всех векторов слов, присутствующих в твите. Длина результирующего вектора будет одинаковой, то есть 200. Мы повторим тот же процесс для всех твитов в наших данных и получим их векторы. Теперь у нас есть 200 функций word2vec для наших данных.__
Необходимо создать вектор для каждого твита, взяв среднее значение векторов слов, присутствующих в твите. В цикле сделать:  vec += model_w2v[word].reshape((1, size)) и поделить финальный вектор на количество слов в твите. 
На выходе должен получиться wordvec_df.shape = (49159, 200).

```python
size = 200
tweet_vec = np.zeros((len(test_df_prepare), size))
mean_vec = []

for n, item in test_df_prepare.iterrows():
    vec = 0
    target_field = item['tweet_token']   
    for word in target_field:
        try:
            vec += modelW2V[word].reshape((1, size))
        except:
            continue
    try:
        mean_vec.append(np.mean(vec, axis=0))
    except:
        mean_vec.append(np.zeros(200))
```

```python
mean_vec = np.array(mean_vec)
mean_vec.shape
```

```python
wordvec_df = pd.DataFrame(mean_vec, columns=['pos_' + str(i+1) for i in range(mean_vec.shape[1])])
```
